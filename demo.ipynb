```python
from dags import *
```

```python
def function_a():
    print("Function A")

def function_b():
    print("Function B")

def function_c():
    print("Function C")

def function_d():
    print("Function D")
```

```python
dag = DAG()

node_A = dag.add_node(function_a)
node_B = dag.add_node(function_b)
node_C = dag.add_node(function_c)
node_D = dag.add_node(function_d)

dag.add_edge(node_A, node_B)
dag.add_edge(node_B, node_C)
dag.add_edge(node_B, node_D)
dag.add_edge(node_C, node_D)

dag.visualize()
```

```python
dag.execute()
```
